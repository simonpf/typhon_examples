```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from typhon.arts import xml
from typhon.plots import profile_p, styles
from typhon.physics import vmr2relative_humidity
plt.style.use(styles.get('typhon'))
```

# Load simulated observations and plot some fields

```python
atm_fields = xml.load("/scratch/uni/u237/users/mprange/phd/iasi_retrieval/iasi_obs_simulation/atm_fields.xml")
iasi_obs = xml.load("/scratch/uni/u237/users/mprange/phd/iasi_retrieval/iasi_obs_simulation/iasi_obs.xml")
```

```python
fig, axs = plt.subplots(ncols=3, sharey=False)
for atm in range(5):
    profile_p(atm_fields[0].grids[1], atm_fields[atm][0][:,0,0], ax=axs[0]) # Temperature
    profile_p(atm_fields[0].grids[1], atm_fields[atm][2][:,0,0], ax=axs[1]) # H2O VMR
    RH = vmr2relative_humidity(atm_fields[atm][2][:,0,0], 
                               atm_fields[0].grids[1], 
                               atm_fields[atm][0][:,0,0])
    profile_p(atm_fields[0].grids[1], RH, ax=axs[2]) # Relative Humidity
axs[0].set_xlabel("Temperature [K]")
axs[1].set_xlabel("H2O VMR [-]")
axs[2].set_xlabel("Relative Humidity [-]")
axs[2].set_xlim([0, 1])
```

# Load retrieved quantities and plot them

```python
retrieved_vmr = xml.load("/scratch/uni/u237/users/mprange/phd/iasi_retrieval/retrieved/vmr_retrieved.xml")
retrieved_T = xml.load("/scratch/uni/u237/users/mprange/phd/iasi_retrieval/retrieved/T_retrieved.xml")
apriori_vmr = xml.load("/scratch/uni/u237/users/mprange/phd/iasi_retrieval/a_priori/vmr_apriori.xml")
apriori_T = xml.load("/scratch/uni/u237/users/mprange/phd/iasi_retrieval/a_priori/T_apriori.xml")
```

```python
fig, axs = plt.subplots(ncols=3, sharey=False)
for atm in range(1):
    profile_p(atm_fields[0].grids[1], retrieved_T[:,0,0], ax=axs[0], label="retrieved") # Temperature
    profile_p(atm_fields[0].grids[1], retrieved_vmr[0,:,0,0], ax=axs[1]) # H2O VMR
    RH = vmr2relative_humidity(retrieved_vmr[0,:,0,0], 
                               atm_fields[0].grids[1], 
                               retrieved_T[:,0,0])
    profile_p(atm_fields[0].grids[1], RH, ax=axs[2]) # Relative Humidity
    
    profile_p(atm_fields[0].grids[1], apriori_T[:,0,0], ax=axs[0], label="a priori") # Temperature
    profile_p(atm_fields[0].grids[1], apriori_vmr[0,:,0,0], ax=axs[1]) # H2O VMR
    RH = vmr2relative_humidity(apriori_vmr[0,:,0,0], 
                               atm_fields[0].grids[1], 
                               apriori_T[:,0,0])
    profile_p(atm_fields[0].grids[1], RH, ax=axs[2]) # Relative Humidity
    
    profile_p(atm_fields[0].grids[1], atm_fields[atm][0][:,0,0], ax=axs[0], label="truth") # Temperature
    profile_p(atm_fields[0].grids[1], atm_fields[atm][2][:,0,0], ax=axs[1]) # H2O VMR
    RH = vmr2relative_humidity(atm_fields[atm][2][:,0,0], 
                               atm_fields[0].grids[1], 
                               atm_fields[atm][0][:,0,0])
    profile_p(atm_fields[0].grids[1], RH, ax=axs[2]) # Relative Humidity
    
axs[0].legend()
axs[1].set_xlabel("H2O VMR [-]")
axs[2].set_xlabel("Relative Humidity [-]")
axs[2].set_xlim([0, 1])
```
